```python
def preproc(df):
    df.DT_REQUISICAO = df.DT_REQUISICAO.fillna(np.nanmean(df['DT_NASCIMENTO']))
    df.DT_NASCIMENTO = df.DT_NASCIMENTO.fillna(np.nanmean(df['DT_NASCIMENTO']))
    df.DS_TIPO_PREST_SOLICITANTE = df.DS_TIPO_PREST_SOLICITANTE.fillna('0')
    df.DS_TIPO_CONSULTA = df.DS_TIPO_CONSULTA.fillna('0')
    df.DS_CBO = df.DS_CBO.fillna('0')
    df.DS_INDICACAO_ACIDENTE = df.DS_INDICACAO_ACIDENTE.fillna('0')
    df.DS_TIPO_INTERNACAO = df.DS_TIPO_INTERNACAO.fillna('0')
    df.DS_TIPO_ACOMODACAO = df.DS_TIPO_ACOMODACAO.fillna('0')
    df.QT_SOLICITADA = df.QT_SOLICITADA.fillna(0)
    df.DS_CARATER_ATENDIMENTO = df.DS_CARATER_ATENDIMENTO.fillna('0')
    df.DS_TIPO_ATENDIMENTO = df.DS_TIPO_ATENDIMENTO.fillna('0')
    df['DT_REQUISICAO'] = pd.to_datetime(df['DT_REQUISICAO'].map(jd_to_date).map(lambda t: '-'.join([str(k) for k in t])), format='%Y-%m-%d')
    df['DT_NASCIMENTO'] = pd.to_datetime(df['DT_NASCIMENTO'].map(jd_to_date).map(lambda t: '-'.join([str(k) for k in t])), format='%Y-%m-%d')
    df['IDADE'] = pd.DatetimeIndex(df['DT_REQUISICAO']).year - pd.DatetimeIndex(df['DT_NASCIMENTO']).year
    return df
```

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder, OneHotEncoder
import numpy as np

def standardScalerFunc(data):
  ss = StandardScaler()
  sstransformed = ss.fit_transform(data)

  return ss, pd.DataFrame(sstransformed)

def labelEncoderFunc(data):
  ss = LabelEncoder()
  sstransformed = ss.fit_transform(data) 

  return ss, pd.DataFrame(sstransformed)

def oneHotEncoderFunc(data):
  ohe = OneHotEncoder(sparse=False, handle_unknown='ignore')
  ohetransformed = ohe.fit_transform(data)
  
  return ohe, pd.DataFrame(ohetransformed)
```

```python
# https://github.com/meshula/exifotio/blob/master/jdutil.py
import math

def jd_to_date(jd):
    jd = jd + 0.5
    
    F, I = math.modf(jd)
    I = int(I)
    
    A = math.trunc((I - 1867216.25)/36524.25)
    
    if I > 2299160:
        B = I + 1 + A - math.trunc(A / 4.)
    else:
        B = I
        
    C = B + 1524
    
    D = math.trunc((C - 122.1) / 365.25)
    
    E = math.trunc(365.25 * D)
    
    G = math.trunc((C - E) / 30.6001)
    
    day = C - E + F - math.trunc(30.6001 * G)
    
    if G < 13.5:
        month = G - 1
    else:
        month = G - 13
        
    if month > 2.5:
        year = D - 4716
    else:
        year = D - 4715
        
    return year, month, math.trunc(day)
```

```python
df_train = preproc(df_train)
numeric = df_train[['QT_SOLICITADA','IDADE']]
categoric = df_train[['DS_TIPO_PREST_SOLICITANTE','DS_CBO','DS_TIPO_CONSULTA', 'DS_INDICACAO_ACIDENTE','DS_CARATER_ATENDIMENTO','DS_TIPO_INTERNACAO','DS_TIPO_ACOMODACAO','DS_TIPO_ATENDIMENTO']]
target = df_train['DS_STATUS_ITEM']

X = pd.concat([numeric, categoric], axis=1)
print(len(X))
X = X.dropna()
print(len(X))
y = target
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, f1_score

#X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0) # 80% training 20% test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y,random_state=1)

#train
ss, x_train_ss = standardScalerFunc(X_train[['QT_SOLICITADA','IDADE']])
ohe, x_train_ohe = oneHotEncoderFunc(X_train[['DS_TIPO_PREST_SOLICITANTE','DS_CBO','DS_TIPO_CONSULTA', 'DS_INDICACAO_ACIDENTE','DS_CARATER_ATENDIMENTO','DS_TIPO_INTERNACAO','DS_TIPO_ACOMODACAO','DS_TIPO_ATENDIMENTO']])

x_train = pd.concat([x_train_ss,x_train_ohe], axis=1)

#test
x_test_ss = ss.transform(X_test[['QT_SOLICITADA','IDADE']])
x_test_ohe = ohe.transform(X_test[['DS_TIPO_PREST_SOLICITANTE','DS_CBO','DS_TIPO_CONSULTA', 'DS_INDICACAO_ACIDENTE','DS_CARATER_ATENDIMENTO','DS_TIPO_INTERNACAO','DS_TIPO_ACOMODACAO','DS_TIPO_ATENDIMENTO']])

x_test = pd.concat([pd.DataFrame(x_test_ss),pd.DataFrame(x_test_ohe)], axis=1)

#Random forest
#classifier = RandomForestClassifier(n_estimators=20, random_state=0)
#classifier.fit(x_train, y_train)
#y_pred = classifier.predict(x_test)

#SVM kernel rbf
#clf = svm.SVC(kernel='rbf')
#clf.fit(x_train, y_train)
#y_pred = clf.predict(x_test)

#Decision tree
#from sklearn.tree import DecisionTreeClassifier
#classifier = DecisionTreeClassifier(random_state=0)
#classifier.fit(x_train, y_train)
#y_pred = classifier.predict(x_test)
```

```python
#MLP sklearn
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import make_classification

classifier = MLPClassifier(random_state=1, max_iter=300).fit(x_train, y_train)
#classifier = MLPClassifier(hidden_layer_sizes=2,random_state=1, max_iter=300, activation='relu', solver='sgd', learning_rate='invscaling').fit(x_train, y_train)
classifier.predict_proba(x_test[:1])

y_pred = classifier.predict(x_test)

classifier.fit(x_train, y_train)

classifier.score(x_test, y_test)
```

```python
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
print(accuracy_score(y_test, y_pred))
```
